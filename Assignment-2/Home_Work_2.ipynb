Question1: Rock-paper-scissors
Steps:
1- Get input from user
2- Use Random module to get random selection
3- convert input to lower case
4- check whether input is integer or character
5- examine the user input with the computer choice

```python
import random
#Create List to hold options [Rock, Paper, Scissors]
option_list=['rock','paper','scissors']
while True:
# Ask user to Enter his selection
    print("To Finish the Game enter (done) Or (DONE)")
    user_input=input("Enter Your Choice: ")
    low_characters=user_input.lower()
    if low_characters =="done":
        print("The game is Finished")
        break
# check whether the input is characters nothing else
    if low_characters.isalpha():
# select random choice from the list
        computer_selection = random.choice(option_list)
        print("Computer Chooses: ",computer_selection)
        if low_characters == computer_selection:
            print("tie!")
        if low_characters != computer_selection:

            if low_characters == "rock" and computer_selection == "scissors":
                print("You,the user, win!")
            if low_characters == "scissors" and computer_selection == "rock":
                print("I, the Computer, win!")

            if low_characters == "scissors" and computer_selection == "paper":
                print("You,the user, win!")
            if low_characters == "paper" and computer_selection == "scissors":
                print("I, the Computer, win!")

            if low_characters == "paper" and computer_selection == "rock":
                print("You,the user, win!")
            if low_characters == "rock" and computer_selection == "paper":
                print("I, the Computer, win!")
            if low_characters !="rock" and low_characters != "scissors" and low_characters !="paper":
                print("No One Win OR Lose")

    else:
        print("Can not enter Digits!")
```

Question4:palindrome
Steps:
1- Get input from user
2- check if input is int or string
3- check each char from begin and end of string if macth return palindrome

```python
print("TO STOP THE PROGRAM PRESS (N) OR (n)")
input_string = input("Enter a string: ")
        
while True:
    count = 0
    if input_string == 'n' or input_string =='N':
        break
    input_string2 = input_string.replace(" ", "").lower()
    if input_string2.isalpha():
    # get middle of string 
        middle = 0 + len(input_string2)//2
# loop from index 0 to half of string
        for i in range(0,middle+1):
# check the i element with end element-i 
            if input_string2[i] != input_string2[len(input_string2)-i-1]:
    # count increase incase there are two chars not match
                count = count +1
    # if all chars are match count will be 0
        if count == 0:
            print("{} is palindrome".format(input_string))
            print("Do you want to enter new string! ")
            input_string = input("Enter a string: ")
            if input_string == 'n' or input_string == 'N':
                break
        else:
            print("{} is not palindrome".format(input_string))
            print("Do you want to enter new string! ")
            input_string = input("Enter a string: ")
            if input_string == 'n' or input_string == 'N':
                break

    else:
        print("You must Enter String noting Else: ")
        input_string = input("Enter a string: ")
```

Question5: File
Steps:
1- Get file name from user and check if name is alpha or integer
2- open file if the name is exist
3- loop throught the file and count the lines,words, characters
4- find Average of words

```python
file_name = input("What is the filename: ")
lines =0
characters=0
words=0
average=0
if file_name.isalpha():

    file_name = file_name.lower()
# return True if the user input same as file name
    if file_name =='robertfrost':
        try:
        # create pointer to the file
            link_file = open(file_name+'.txt','r')
            for i in link_file:
                no_space = i.strip()
                lines = lines+1

                str_count = len(i.split())
                words = words+str_count

                for j in no_space:
                    if j != ' ':
                        characters=characters+1

            average = characters/words
            link_file.close()
        except:
            print("FILE NOT OPENED")

        print("Number of lines: ",lines)
        print("Number of words: ", words)

        print("Number of characters: ",characters)
        print("Average length of a word: {:.1f}".format(average))


    else:
        print("File name not exist")
else:
    print("FILE NAME Must Contain String!")
```
