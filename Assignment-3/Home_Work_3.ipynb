Question 1:
1- Get file name from user then open file and read 
2- pull out all lines as Key , value and store it in Dictionary
3- Mix the characters of random key and print it in screen
4- get user guess and compare it with the scramble word
5- select ? in case of meaning 
6- finish the program 'no' or continue 'yes'

```python
import random

file_name = input("Give the name of the “words and their meanings” file: ")
dictionary = dict()
file_name = file_name.lower()
try:
            link_file = open(file_name)
            print("Unscramble the following letters to form a word. Type “?” for the meaning of")
    # Loop throught the line by line
            for i in link_file:
    # return Key: value of each line that seprate by Comma
                key, value = i.strip().split(',')
    # Assing Value to each Key.
                dictionary[key]=value
            flag = True

        # return List of Keys
            list_dictionary = dictionary.keys()
            list_dictionary = list(list_dictionary)


            while flag:
                flag2 = True
                scramble_word = ''
                word_list = random.choice(list_dictionary)  # select random key from the list
                correct_word = word_list
                new_word = word_list[1:-1]
                new_word = random.sample(new_word, len(new_word))
                new_word.insert(0, word_list[0])
                new_word.append(word_list[-1])
                scramble_word = scramble_word + ''.join(new_word) + " "

                print("the unscrambled word:",scramble_word)
                while flag2:
                    user_input = input("Enter the answer [or ? for the meaning]: ")
                    if user_input == '?':
                        print("The word means:",dictionary[correct_word])

                    elif user_input != '?':
                         if user_input == correct_word:
                             user_input= input("You got it! Do you want to continue [yes or no]:")

                             while True:
                                if user_input.lower() == 'yes':
                                    flag2 =False
                                    break
                                elif user_input.lower() =='no':
                                    print("Goodbye!")
                                    flag = False
                                    flag2 = False
                                    break
                                else:
                                   user_input = input("You Should Enter yes Or no: ").lower()
                         else:
                           print("Wrong, try again ")
            link_file.close()
except:
            print("FILE NOT OPENED")
            print("File name not exist")
```
